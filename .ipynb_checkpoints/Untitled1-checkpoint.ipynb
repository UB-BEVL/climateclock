```python
# === COLUMN NAMES (exactly as you provided) ===
EPW_COLUMNS = [
    "year","month","day","hour","minute","datasource",
    "drybulb","dewpoint","relhum","atmos_pressure",
    "exthorrad","extdirrad","horirsky",
    "glohorrad","dirnorrad","difhorrad",
    "glohorillum","dirnorillum","difhorillum","zenlum",
    "winddir","windspd",
    "totskycvr","opaqskycvr",
    "visibility","ceiling_hgt",
    "presweathobs","presweathcodes",
    "precip_wtr","aerosol_opt_depth",
    "snowdepth","days_last_snow",
    "albedo","liq_precip_depth","liq_precip_rate"
]

# === UNITS (exactly as you provided) ===
EPW_UNITS = {
    "year":"-", "month":"-", "day":"-", "hour":"-", "minute":"-", "datasource":"-",
    "drybulb":"C", "dewpoint":"C", "relhum":"%", "atmos_pressure":"Pa",
    "exthorrad":"Wh/m2", "extdirrad":"Wh/m2", "horirsky":"Wh/m2",
    "glohorrad":"Wh/m2", "dirnorrad":"Wh/m2", "difhorrad":"Wh/m2",
    "glohorillum":"lux", "dirnorillum":"lux", "difhorillum":"lux", "zenlum":"lux",
    "winddir":"degrees", "windspd":"m/s",
    "totskycvr":"tenths", "opaqskycvr":"tenths",
    "visibility":"km", "ceiling_hgt":"m",
    "presweathobs":"-", "presweathcodes":"-",
    "precip_wtr":"mm", "aerosol_opt_depth":"thousandths",
    "snowdepth":"cm", "days_last_snow":"-",
    "albedo":"-", "liq_precip_depth":"mm", "liq_precip_rate":"hour",
}

# === a simple dtype plan: ints for Y/M/D/H/min; floats for meteorology; strings for codes ===
NUMERIC_FLOAT = {
    "drybulb","dewpoint","relhum","atmos_pressure",
    "exthorrad","extdirrad","horirsky","glohorrad","dirnorrad","difhorrad",
    "glohorillum","dirnorillum","difhorillum","zenlum",
    "winddir","windspd","totskycvr","opaqskycvr","visibility","ceiling_hgt",
    "precip_wtr","aerosol_opt_depth","snowdepth","days_last_snow","albedo",
    "liq_precip_depth","liq_precip_rate"
}
STRINGY = {"datasource","presweathobs","presweathcodes"}
```

```python
from dataclasses import dataclass, asdict
from pathlib import Path
import csv, re
import pandas as pd

@dataclass
class EPWProvenance:
    source: str|None
    station_wmo: str|None
    file_kind: str|None     # TMY3 / TMYx / etc.
    years_in_file: list[int]
    filename: str
    city: str|None
    country: str|None

def _parse_location(location_line: str) -> dict:
    # LOCATION,City,State,Country,Source,WMO,Lat,Lon,Timezone,Elevation
    parts = [p.strip() for p in location_line.split(",")]
    parts += [""] * max(0, 10 - len(parts))
    def fnum(x):
        try: return float(x) if x != "" else None
        except: return None
    return {
        "city": parts[1] if len(parts)>1 else "",
        "state_province": parts[2] if len(parts)>2 else "",
        "country": parts[3] if len(parts)>3 else "",
        "source": parts[4] if len(parts)>4 else "",
        "wmo": parts[5] if len(parts)>5 else "",
        "latitude": fnum(parts[6]) if len(parts)>6 else None,
        "longitude": fnum(parts[7]) if len(parts)>7 else None,
        "timezone": fnum(parts[8]) if len(parts)>8 else None,
        "elevation_m": fnum(parts[9]) if len(parts)>9 else None,
    }

def _guess_kind_and_years(filename: str, comments1: str, comments2: str):
    text = f"{filename} {comments1} {comments2}"
    kind = "TMY3" if "TMY3" in text else ("TMYx" if "TMYx" in text else ("TMY" if "TMY" in text else None))
    years = sorted({int(y) for y in re.findall(r"(19|20)\d{2}", text)})
    return kind, years

def read_epw_with_schema(path: str|Path):
    """
    Returns (header_dict, df, provenance) using YOUR column names.
    - df has hourly rows with a normalized timestamp index (EPW 1–24 -> 0–23).
    """
    path = Path(path)
    lines = path.read_text(encoding="latin-1", errors="replace").splitlines()
    if len(lines) < 9:
        raise ValueError("Not a valid EPW (needs 8 header lines + data).")

    header_lines, data_lines = lines[:8], lines[8:]

    # Header object (we keep it small & readable for now)
    loc = _parse_location(header_lines[0])
    header = {
        "location": loc,
        "design_conditions": header_lines[1],
        "typical_extreme_periods": header_lines[2],
        "ground_temps": header_lines[3],
        "holidays_dst": header_lines[4],
        "data_periods": header_lines[5],
        "comments1": header_lines[6] if len(header_lines)>6 else "",
        "comments2": header_lines[7] if len(header_lines)>7 else "",
    }

    # Hourly data
    rows = list(csv.reader(data_lines))
    df = pd.DataFrame(rows)
    # Align incoming columns to your schema length
    df = df.iloc[:, :len(EPW_COLUMNS)]
    df.columns = EPW_COLUMNS

    # Coerce types
    df["year"]   = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
    df["month"]  = pd.to_numeric(df["month"], errors="coerce").astype("Int64")
    df["day"]    = pd.to_numeric(df["day"], errors="coerce").astype("Int64")
    df["hour"]   = pd.to_numeric(df["hour"], errors="coerce").astype("Int64")
    df["minute"] = pd.to_numeric(df["minute"], errors="coerce").astype("Int64")

    for c in EPW_COLUMNS:
        if c in {"year","month","day","hour","minute"}: 
            continue
        if c in STRINGY:
            # keep as string/object
            df[c] = df[c].astype(str)
        else:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    # Normalize time index: EPW hour is END-of-hour (1..24) → represent START-of-hour
    ts = pd.to_datetime(
        dict(year=df["year"], month=df["month"], day=df["day"], hour=df["hour"]),
        errors="coerce"
    ) - pd.to_timedelta(1, "h")
    df.index = ts
    df.index.name = "timestamp"

    # Provenance (simple but useful)
    kind, years = _guess_kind_and_years(path.name, header["comments1"], header["comments2"])
    years_in_file = years or sorted(set(int(x) for x in df["year"].dropna().tolist()))
    prov = EPWProvenance(
        source=loc.get("source") or "OneBuilding",
        station_wmo=loc.get("wmo") or None,
        file_kind=kind,
        years_in_file=years_in_file,
        filename=path.name,
        city=loc.get("city"), country=loc.get("country"),
    )
    return header, df, prov
```

```python
from pathlib import Path
import requests, zipfile, os

zip_url = "https://climate.onebuilding.org/WMO_Region_4_North_and_Central_America/USA_United_States_of_America/NY_New_York/USA_NY_CGS.Buffalo.725285_TMYx.zip"

# download
out_dir = Path("weather_cache"); out_dir.mkdir(exist_ok=True)
zip_path = out_dir / zip_url.split("/")[-1]
with requests.get(zip_url, stream=True, timeout=60, headers={"User-Agent":"Mozilla/5.0"}) as r:
    r.raise_for_status()
    with open(zip_path, "wb") as f:
        for b in r.iter_content(1<<14):
            if b: f.write(b)

# extract a .epw
with zipfile.ZipFile(zip_path, "r") as z:
    epws = [m for m in z.namelist() if m.lower().endswith(".epw")]
    epws.sort(key=lambda m: z.getinfo(m).file_size, reverse=True)
    member = epws[0]
    epw_path = out_dir / Path(member).name
    z.extract(member, path=out_dir)
    extracted = out_dir / member
    if extracted != epw_path:
        if epw_path.exists(): epw_path.unlink()
        os.replace(extracted, epw_path)

# ingest
header, df, prov = read_epw_with_schema(epw_path)
header["location"], df.shape, (df.index.min(), df.index.max()), prov

```

```python
import re
import pandas as pd

def parse_header_metadata(header: dict) -> dict:
    loc = header.get("location", {}) or {}
    periods_raw = header.get("data_periods", "") or ""
    # heuristic parse; store raw if no match
    m = re.findall(
        r"DATA\s*PERIODS\s*,\s*(\d+)\s*,\s*(\d+)\s*,\s*([A-Za-z]+)\s*,\s*([A-Za-z]+)\s*,\s*([0-9/ ]+)\s*,\s*([0-9/ ]+)",
        periods_raw, flags=re.I
    )
    periods = [{"raw": periods_raw}] if not m else [
        {"period_index": int(pi), "data_type": dt, "first_day_of_week": dow,
         "start_md": re.sub(r"\s+","", s), "end_md": re.sub(r"\s+","", e)}
        for _, pi, dt, dow, s, e in m
    ]
    hol_raw = header.get("holidays_dst", "") or ""
    dst = {"raw": hol_raw}
    if re.search(r"\bYes\b", hol_raw, flags=re.I):
        dst["dst_observed"] = True
        mstart = re.search(r"Start\s*,\s*([^,]+)", hol_raw, flags=re.I)
        mend   = re.search(r"End\s*,\s*([^,]+)", hol_raw, flags=re.I)
        if mstart: dst["dst_start"] = mstart.group(1).strip()
        if mend:   dst["dst_end"] = mend.group(1).strip()
    elif re.search(r"\bNo\b", hol_raw, flags=re.I):
        dst["dst_observed"] = False

    return {
        "location": {
            "city": loc.get("city"),
            "state_province": loc.get("state_province"),
            "country": loc.get("country"),
            "latitude": loc.get("latitude"),
            "longitude": loc.get("longitude"),
            "timezone": loc.get("timezone"),
            "elevation_m": loc.get("elevation_m"),
            "wmo": loc.get("wmo"),
            "source": loc.get("source"),
        },
        "data_periods": periods,
        "dst": dst,
    }

def completeness_report(df: pd.DataFrame) -> pd.DataFrame:
    total = len(df)
    miss = df.isna().sum()
    filled = total - miss
    pct_missing = (miss / total * 100).round(2)
    return pd.DataFrame({"filled": filled.astype(int),
                         "missing": miss.astype(int),
                         "pct_missing": pct_missing})

def time_audit(df: pd.DataFrame) -> dict:
    if not isinstance(df.index, pd.DatetimeIndex) or df.index.tz is not None:
        return {"time_index_ok": False, "reason": "Index is not a naive DatetimeIndex"}
    start, end = df.index.min(), df.index.max()
    expected = pd.date_range(start=start, end=end, freq="H")
    dup_count = int(pd.Index(df.index).duplicated().sum())
    missing_ts = expected.difference(df.index)
    return {
        "time_index_ok": True,
        "coverage": {"start": str(start), "end": str(end)},
        "expected_hours": len(expected),
        "actual_rows": len(df),
        "missing_timestamps": int(len(missing_ts)),
        "duplicate_timestamps": dup_count,
    }

DEFAULT_RANGES = {
    "drybulb": (-60, 60), "dewpoint": (-80, 50), "relhum": (0, 100),
    "atmos_pressure": (80000, 110000),
    "glohorrad": (0, 1400), "dirnorrad": (0, 1400), "difhorrad": (0, 1400),
    "exthorrad": (0, 1500), "extdirrad": (0, 1500),
    "windspd": (0, 100), "winddir": (0, 360),
}

def sanity_checks(df: pd.DataFrame, ranges: dict = None) -> pd.DataFrame:
    ranges = ranges or DEFAULT_RANGES
    recs = []
    for col, (lo, hi) in ranges.items():
        if col in df.columns:
            s = df[col]
            recs.append({
                "column": col, "range_lo": lo, "range_hi": hi,
                "n_total": len(s), "n_nan": int(s.isna().sum()),
                "n_below_lo": int((s < lo).sum()),
                "n_above_hi": int((s > hi).sum()),
                "n_out_of_range": int(((s < lo) | (s > hi)).sum()),
            })
    return pd.DataFrame(recs).sort_values("column")

# Run the audit
metadata = parse_header_metadata(header)
completeness = completeness_report(df)
timecheck = time_audit(df)
sanity = sanity_checks(df)

metadata, timecheck, completeness.head(), sanity.head()
```

```python
import numpy as np
import pandas as pd

# ---- minimal preprocess
wx = df.copy()

# short-gap interpolate a few continuous vars (≤3h)
interp_cols = [
    "drybulb","dewpoint","relhum","atmos_pressure",
    "windspd","winddir",
    "glohorrad","dirnorrad","difhorrad","exthorrad","extdirrad"
]
for c in interp_cols:
    if c in wx.columns:
        wx[c] = wx[c].interpolate(limit=3, limit_direction="both")

# clip RH to [0,100]
if "relhum" in wx.columns:
    wx["relhum"] = wx["relhum"].clip(0, 100)

# ---- derived (safe fallbacks if packages missing)
T = wx["drybulb"] if "drybulb" in wx else pd.Series(index=wx.index, dtype=float)
RH = wx["relhum"]  if "relhum"  in wx else pd.Series(index=wx.index, dtype=float)

# saturation vapor pressure (Pa) – Tetens
es = 610.94 * np.exp(17.625 * T / (T + 243.04))
wx["sat_press"] = es
wx["vap_press"] = es * (RH/100.0)
wx["abs_hum"]   = 216.7 * wx["vap_press"] / (T + 273.15)

# wet-bulb (Stull) – requires T & RH
wx["twb"] = (
    T*np.arctan(0.151977*np.sqrt(RH+8.313659))
    + np.arctan(T+RH) - np.arctan(RH-1.676331)
    + 0.00391838*(RH**1.5)*np.arctan(0.023101*RH) - 4.686035
)

# UTCI (optional)
try:
    from pythermalcomfort.models import utci
    v = wx["windspd"] if "windspd" in wx.columns else pd.Series(0, index=wx.index, dtype=float)
    wx["utci"] = utci(tdb=T, tr=T, v=v.fillna(0), rh=RH)
except Exception:
    pass  # skip if package not installed

# solar (optional; skip if pvlib missing)
try:
    import pvlib
    lat = header["location"]["latitude"]; lon = header["location"]["longitude"]; tz = header["location"]["timezone"] or 0
    from datetime import timedelta, timezone
    fixed_tz = timezone(timedelta(hours=tz))
    wx_tz = wx.copy(); wx_tz.index = wx_tz.index.tz_localize(fixed_tz)
    sol = pvlib.solarposition.get_solarposition(wx_tz.index, latitude=lat, longitude=lon)
    wx["elevation"] = sol["elevation"].to_numpy()
    wx["azimuth"]   = sol["azimuth"].to_numpy()
except Exception:
    pass

# ---- assemble Clima DataFrame
cols_base = [
    "drybulb","dewpoint","relhum","atmos_pressure",
    "exthorrad","extdirrad","horirsky",
    "glohorrad","dirnorrad","difhorrad",
    "glohorillum","dirnorillum","difhorillum","zenlum",
    "winddir","windspd","totskycvr","opaqskycvr","visibility"
]
cols_derived = ["utci","vap_press","abs_hum","twb","sat_press","elevation","azimuth"]
keep = [c for c in cols_base + cols_derived if c in wx.columns]

cdf = wx[keep].copy()
cdf["month"] = cdf.index.month
cdf["day"]   = cdf.index.day
cdf["hour"]  = cdf.index.hour
cdf["doy"]   = cdf.index.dayofyear

cdf.head(3)
```

```python
# === Cell 5: Temperature & Humidity visuals ===
import pandas as pd
import matplotlib.pyplot as plt

# 5a) Annual trends: daily mean lines
daily = cdf.resample("D").mean(numeric_only=True)

plt.figure()
daily["drybulb"].plot()
plt.title("Annual Dry-Bulb Temperature (Daily Mean)")
plt.xlabel("Date"); plt.ylabel("°C")
plt.show()

plt.figure()
daily["relhum"].plot()
plt.title("Annual Relative Humidity (Daily Mean)")
plt.xlabel("Date"); plt.ylabel("%")
plt.show()

# 5b) Daily scatter per month (hour on x-axis)
# Dry-bulb
for m in range(1, 13):
    subset = cdf[cdf["month"] == m]
    if subset.empty: 
        continue
    plt.figure()
    plt.scatter(subset["hour"], subset["drybulb"], s=6, alpha=0.4)
    plt.title(f"Dry-Bulb — Daily Hourly Scatter (Month {m})")
    plt.xlabel("Hour"); plt.ylabel("°C")
    plt.show()

# Relative Humidity
for m in range(1, 13):
    subset = cdf[cdf["month"] == m]
    if subset.empty:
        continue
    plt.figure()
    plt.scatter(subset["hour"], subset["relhum"], s=6, alpha=0.4)
    plt.title(f"Relative Humidity — Daily Hourly Scatter (Month {m})")
    plt.xlabel("Hour"); plt.ylabel("%")
    plt.show()

# 5c) Heatmaps (Day-of-Year × Hour)
def heat_matrix(series: pd.Series) -> pd.DataFrame:
    tmp = pd.DataFrame({
        "doy": series.index.dayofyear, 
        "hour": series.index.hour, 
        "val": series.values
    }).dropna()
    mat = tmp.pivot_table(index="doy", columns="hour", values="val", aggfunc="mean")
    return mat.sort_index()

# Dry-bulb heatmap
mat_t = heat_matrix(cdf["drybulb"])
plt.figure()
plt.imshow(mat_t.values, aspect="auto", origin="lower")
plt.title("Dry-Bulb Temperature Heatmap (Day-of-Year × Hour)")
plt.xlabel("Hour"); plt.ylabel("Day of Year")
plt.colorbar(label="°C")
plt.show()

# RH heatmap
mat_rh = heat_matrix(cdf["relhum"])
plt.figure()
plt.imshow(mat_rh.values, aspect="auto", origin="lower")
plt.title("Relative Humidity Heatmap (Day-of-Year × Hour)")
plt.xlabel("Hour"); plt.ylabel("Day of Year")
plt.colorbar(label="%")
plt.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# --- 1) Collapse to a canonical calendar year (handles files with multiple years) ---
# Build a "calendar day" index with a fixed year so Jan→Dec is continuous
cal = cdf.copy()
cal["month"] = cal.index.month
cal["day"]   = cal.index.day
cal_date = pd.to_datetime(
    dict(year=2001, month=cal["month"], day=cal["day"]),
    errors="coerce"
)

# Daily aggregates across ALL years for each calendar day
daily = (
    cal.assign(_caldate=cal_date)
       .groupby("_caldate")
       .agg(drybulb_mean=("drybulb", "mean"),
            drybulb_min =("drybulb", "min"),
            drybulb_max =("drybulb", "max"))
       .sort_index()
)

# 7-day rolling mean for a smoother "average dry-bulb" line
daily["drybulb_mean_smooth"] = daily["drybulb_mean"].rolling(7, min_periods=1, center=True).mean()

# --- 2) ASHRAE 55 adaptive comfort bands based on running-mean outdoor temperature ---
# Prevailing mean outdoor temp (Trm) ~ exponentially weighted mean of daily means
alpha = 0.8
Trm = daily["drybulb_mean"].ewm(alpha=(1-alpha), adjust=False).mean()

# T_comf = 0.31*Trm + 17.8  (valid for ~10–33.5°C Trm); clamp to reasonable range
T_comf = (0.31 * Trm + 17.8).clip(lower=-10, upper=40)
c80_lo, c80_hi = T_comf - 2.5, T_comf + 2.5
c90_lo, c90_hi = T_comf - 3.5, T_comf + 3.5

# --- 3) Build the interactive figure (clean, Clima-like styling) ---
x = daily.index

fig = go.Figure()

# Min/Max ribbon (light red)
fig.add_trace(go.Scatter(
    x=x, y=daily["drybulb_max"], name="Dry-bulb range (max)",
    mode="lines", line=dict(width=0), hoverinfo="skip", showlegend=False
))
fig.add_trace(go.Scatter(
    x=x, y=daily["drybulb_min"], name="Dry-bulb range (min)",
    mode="lines", line=dict(width=0),
    fill="tonexty", fillcolor="rgba(255, 0, 0, 0.18)",
    hoverinfo="skip", showlegend=True
))

# Smooth average (red line)
fig.add_trace(go.Scatter(
    x=x, y=daily["drybulb_mean_smooth"], name="Average dry-bulb",
    mode="lines",
    line=dict(width=2.5, color="rgb(220, 20, 60)"),
    hovertemplate="Date: %{x|%b %d}<br>Avg (7-day): %{y:.1f} °C<extra></extra>"
))

# Comfort ribbons (gray, behind the line)
fig.add_trace(go.Scatter(
    x=x, y=c90_hi, name="ASHRAE adaptive (90%)",
    mode="lines", line=dict(width=0), hoverinfo="skip", showlegend=True
))
fig.add_trace(go.Scatter(
    x=x, y=c90_lo, name="ASHRAE adaptive (90%)",
    mode="lines", line=dict(width=0), fill="tonexty",
    fillcolor="rgba(120,120,120,0.12)", hoverinfo="skip", showlegend=False
))
fig.add_trace(go.Scatter(
    x=x, y=c80_hi, name="ASHRAE adaptive (80%)",
    mode="lines", line=dict(width=0), hoverinfo="skip", showlegend=True
))
fig.add_trace(go.Scatter(
    x=x, y=c80_lo, name="ASHRAE adaptive (80%)",
    mode="lines", line=dict(width=0), fill="tonexty",
    fillcolor="rgba(160,160,160,0.22)", hoverinfo="skip", showlegend=False
))

# Layout: Jan→Dec axis with range slider
fig.update_layout(
    title="Annual Dry-Bulb Temperature — Interactive (Clima style)",
    xaxis_title="", yaxis_title="Dry-bulb temperature (°C)",
    hovermode="x unified",
    legend=dict(orientation="h", yanchor="bottom", y=1.02, x=0),
    margin=dict(l=50,r=20,t=60,b=40),
    xaxis=dict(
        tickformat="%b",
        dtick="M1",
        range=[pd.Timestamp("2001-01-01"), pd.Timestamp("2001-12-31")],
        rangeslider=dict(visible=True),
        type="date",
        rangeselector=dict(
            buttons=[
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=3, label="3m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(step="all")
            ]
        )
    )
)

fig.show()
```
